Question 1:- A digital payment platform recorded the following transaction amounts made by a customer during a single day.

    transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]

    Write a Python program that performs the following analysis:

        1)Identify and display all transaction values that appear more than once.

        2)Detect transactions greater than ₹5000 as these may require fraud monitoring.
        
        3)Calculate the total spending for the day.

        4)Create a new list containing only unique transaction values.

        5)Sort the unique transaction list in descending order to view the largest transactions first.

```python
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
sum=0
fraud_transactions=[]
unique_list=[]
duplicate_values=[]
for i in transactions:
    sum+=i
    if i>5000:
       fraud_transactions.append(i)
    if i not in unique_list:
        unique_list.append(i)
    elif i not in duplicate_values:
        duplicate_values.append(i)
unique_list.sort(reverse=True)
    
print('orignal list:',transactions)
print('The sum of all the transactions is:',sum)
print('These values appear more than once',duplicate_values)
print('These transactions require fraud monitoring:',fraud_transactions )
print('removing duplicate values from the list and sorting list in descending order:',unique_list)
```

Question 2:- A dataset exported from a database contains the following customer age values.

    ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46] Some entries are invalid because they are either negative numbers or missing values (None).

        Write a Python program that:

        1)Iterates through the dataset and removes all invalid age values.

        2)Stores only valid ages in a new list.
        
        3)Calculates the average age of valid customers.
        
        4)Displays the list of customers whose age is greater than 30.
        
        5)Prints how many invalid entries were removed during the cleaning process.


```python
ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]

valid_ages = []
Elder_ages=[]
invalid_count=0
total=0
avg=0
for age in ages:
    if age is not None and age >= 0:
        valid_ages.append(age)
        total+=age
    if age is not None and age>30:
        Elder_ages.append(age)


avg=total/len(valid_ages)
invalid_count=len(ages)-len(valid_ages)



print("Valid ages:", valid_ages)
print('average of all ages:',avg)  
print('Ages which are greater than 30:',Elder_ages) 
print('invalid entries were removed during the cleaning process:',invalid_count) 
```

Question 4:- A web platform records the hours when users log into the system.

    login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10] Write a Python program that:

        1)Separates login hours into even-hour logins and odd-hour logins. 

        2)Identifies login hours that fall into peak activity time (hours greater than 5).

        3)Sorts the login hours in ascending order.

        4)Calculates how many logins occurred during peak hours.

```python
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
even_hours=[]
odd_hours=[]
greater_hours=[]
login_hours.sort()
count=0
for hours in login_hours:
    if hours%2==0:
        even_hours.append(hours)
    else:
        odd_hours.append(hours)
    if hours>5:
        greater_hours.append(hours)
        count+=1




print('Even Hours:',even_hours)
print('Odd Hours:',odd_hours)
print('Login hours that fall into peak activity time (hours greater than 5):',greater_hours)
print('Total logins occurred during peak hours:',count)
```
