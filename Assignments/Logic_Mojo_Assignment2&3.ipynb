1. Customer Segmentation Engine
You are given a list of customer dictionaries:

customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]
Task:

Write a function segment_customer(customer):
If spend divided by income is greater than 0.6 return "High Value"
If spend divided by income is between 0.3 and 0.6 return "Medium Value"
Otherwise return "Low Value"
Create a final dictionary in the format {customer_id: segment}

```python
def segment_customer(customer):
    ratio = customer['spend'] / customer['income']
    
    if ratio > 0.6:
        return "High Value"
    elif 0.3 <= ratio <= 0.6:
        return "Medium Value"
    else:
        return "Low Value"
result = {}

customers = [ {"id": 1, "age": 25, "income": 50000, "spend": 20000}, {"id": 2, "age": 45, "income": 120000, "spend": 80000}, {"id": 3, "age": 35, "income": 70000, "spend": 30000} ] 

for c in customers:
    result[c["id"]] = segment_customer(c)

print(result)
```

Q2. Text Cleaning Pipeline
You are given:

texts = [
    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]
Task:

Write a function clean_text(text):
Remove special characters
Remove extra spaces
Convert text to lowercase
Apply the function to all elements using loop or map
Return cleaned list

```python
import re
texts = [ "Hello!!! How are you???", "This is AI class.", "Python@@@is###great" ] 

def clean_text(text):
    text= re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text=text.lower().strip()
    return text

cleaned=[]
for t in texts:
    text=clean_text(t)
    cleaned.append(text)
cleaned
```

Use lambda with filter or map
Return only fraud transactions

Q4)You are given:

data = [
    {"salary": 50000, "experience": 2},
    {"salary": 100000, "experience": 5}
]
Task:

Write a function feature_engineer(record)
Add a new key salary_per_exp
salary_per_exp is salary divided by experience
Handle division by zero using conditional logic
Return updated dataset

```python
data = [
    {"salary": 50000, "experience": 2},
    {"salary": 100000, "experience": 5}
]

def feature_engineer(dataset):
    updated_data = []

    for record in dataset:
        if record["experience"] != 0:
            record["salary_per_exp"] = record["salary"] / record["experience"]
        else:
            record["salary_per_exp"] = 0   # to avoid division by zero

        updated_data.append(record)

    return updated_data


result = feature_engineer(data)
print(result)



```

6. Sequence Pattern Detection
You are given:

data = [10, 20, 30, 25, 40, 50]
Task:

Write a function detect_increasing_sequence(data)
Find the longest continuous increasing sequence
Return that sequence as a list
Example output:

[10, 20, 30]

```python
data = [10, 20, 30, 25, 40, 50]

longest_list = []
current = [data[0]]

for i in range(1, len(data)):   # ✅ start from 1
    if data[i] > data[i - 1]:
        current.append(data[i])
    else:
        if len(current) > len(longest_list):
            longest_list = current
        current = [data[i]]

print(longest_list)
print(current)
```

7. Dictionary-Based Recommendation System
You are given:

user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}
Task:

Write a function recommend(user)

Recommend items based on common items between users

Do not recommend items already purchased by the user

Use sets and dictionary iteration

```python
def recommend(user, user_purchases):
    user_items = set(user_purchases[user])
    recommendations = set()

    for other_user, items in user_purchases.items():
        if other_user == user:
            continue

        other_items = set(items)
        if user_items & other_items:
            recommendations |= other_items
    return list(recommendations - user_items)
    
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}

print(recommend("user1", user_purchases))
```

8. Recursive Decision Tree Simulation
Task:

Write a recursive function decision_tree(score)

If score is greater than 80 return "Approved"

If score is between 50 and 80 call the function again with score plus 10

If score is less than 50 return "Rejected"

```python
def decision_tree(score):
    if score > 80:
        return "Approved"
    elif score < 50:
        return "Rejected"
    else:
        return decision_tree(score + 10)

score1=int(input('enter score:'))
ans=decision_tree(score1)
print(ans) 
```
